```python
import pandas as pd
```

```python
full_matrix = pd.read_csv('2_full_matrix.csv', index_col=0)

# Drop specified columns
dropColumns = ['BDSPPatientID', 'NoteDate', 'NoteTitle', 'text', 'hospital', 'annot', 'ICD']
keywordMatrix = full_matrix.drop(columns=dropColumns)

keywordMatrix.head()
```

```python
# Separate column names into posKeywords and negKeywords
pos_keywords = [col for col in keywordMatrix.columns if not col.startswith('neg_')]
neg_keywords = [col[4:] for col in keywordMatrix.columns if col.startswith('neg_')]

# Create a DataFrame from the keywords
max_len = max(len(pos_keywords), len(neg_keywords))
pos_keywords.extend([''] * (max_len - len(pos_keywords)))
neg_keywords.extend([''] * (max_len - len(neg_keywords)))

keywords_df = pd.DataFrame({
    'posKeywords': pos_keywords,
    'negKeywords': neg_keywords
})

keywords_df.head()
```

```python
# Create sets for comparison
pos_keywords_set = set(pos_keywords)
neg_keywords_set = set(neg_keywords)

# Find the words that are in posKeywords but not in negKeywords, and vice versa
pos_not_in_neg = pos_keywords_set - neg_keywords_set
neg_not_in_pos = neg_keywords_set - pos_keywords_set

print("Words in posKeywords but not in negKeywords:")
print(pos_not_in_neg)

print("\nWords in negKeywords but not in posKeywords:")
print(neg_not_in_pos)
```
