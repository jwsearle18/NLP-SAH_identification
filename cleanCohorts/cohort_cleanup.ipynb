```python
import pandas as pd
```

## Goal is to cleanup the cohort data
 - Make sure each cohort df has the same column info

```python
BI_neg = pd.read_csv('../annotationToolNegBIDMC/BIDMC_icd_neg_cohort.csv')
BI_pos = pd.read_csv('../annotationToolPosBIDMC/BIDMC_icd_pos_cohort.csv')
MGB_neg = pd.read_csv('../annotationToolNegMGB/MGB_icd_neg_cohort.csv')
MGB_pos = pd.read_csv('../annotationToolPosMGB/MGB_icd_pos_cohort.csv')
```

```python
BI_neg.head()
```

```python
dropColumns = ['ICD', 'AdmissionDate']

BI_neg = BI_neg.drop(columns=dropColumns)
BI_neg.head()
```

```python
BI_pos.head()
```

```python
dropColumns = ['ICD', 'AdmissionDate']

BI_pos = BI_pos.drop(columns=dropColumns)
BI_pos.head()
```

```python
MGB_neg.head()
```

```python
MGB_pos.head()
```

```python
dropColumns = ['ICD', 'DateICD', 'valid']
renameColumns = {
    'DateNote': 'NoteDate',
    'NoteTextFile': 'NoteTitle'
}

MGB_pos = MGB_pos.drop(columns=dropColumns)
MGB_pos = MGB_pos.rename(columns=renameColumns)

MGB_pos.head()
```

```python
# create combined df
combined_df = pd.concat([BI_neg, BI_pos, MGB_neg, MGB_pos], ignore_index=True)
print(len(combined_df))
combined_df.head()
```

```python
# df for each hospital
BI_cohort = pd.concat([BI_neg, BI_pos], ignore_index=True)
print(len(BI_cohort))
BI_cohort.head()

MGB_cohort = pd.concat([MGB_neg, MGB_pos], ignore_index=True)
print(len(MGB_cohort))
MGB_cohort.head()
```

```python
BI_neg.to_csv('BIDMC_neg_cohort_final.csv', index=False)
BI_pos.to_csv('BIDMC_pos_cohort_final.csv', index=False)
MGB_neg.to_csv('MGB_neg_cohort_final.csv', index=False)
MGB_pos.to_csv('MGB_pos_cohort_final.csv', index=False)
combined_df.to_csv('full_cohort_final.csv', index=False)
```

```python
BI_cohort.to_csv('BIDMC_cohort_final.csv', index=False)
MGB_cohort.to_csv('MGB_cohort_final.csv', index=False)
```
